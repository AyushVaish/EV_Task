```python
df.head()

def clean(df,n):

    df=df.copy()
    for i in n:
        q1=df[i].quantile(0.25)
        q3=df[i].quantile(0.75)

        iqr = q3-q1

        lb = q1 - 1.5*iqr
        up = q3 + 1.5*iqr

        df = df[ (df[i] > lb) & (df[i] < up) ]

    return df

#df = clean(df, n = ["AccelSec",'TopSpeed_KmH','Range_Km','Efficiency_WhKm','FastCharge_KmH'])
#df.reset_index(drop=True,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x = df_final.drop(columns = ["PowerTrain"])

y = df_final['PowerTrain']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)

x_test = scaler.fit_transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
```

```python
model_lr.fit(x_train,y_train)
yhat = model_lr.predict(x_test)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,yhat, average='micro')
```

**Model giving 100% accuracy on LogisticRegression. Not a very big surprise as amount of test Data is less.**

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
```

```python
input_dim = x_train.shape[1]
from sklearn.preprocessing import LabelEncoder

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2)
label_encoder = LabelEncoder()

y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)
```

```python
y_train = tf.keras.utils.to_categorical(y_train, num_classes=3)

model_nn = Sequential([
    Dense(32, input_dim = input_dim , activation = "relu"),
    Dense(8, activation = "relu"),
    Dense(3,activation="softmax")
])


model_nn.compile(optimizer='adam', loss='categorical_crossentropy')


model_nn.summary()
```

```python
model_nn.fit(x_train,y_train , verbose=0 , epochs = 50)
```

```python
yhat_nn = model_nn.predict(x_test)

yhat_nn_classes = np.argmax(yhat_nn, axis=1)

f1_nn = f1_score(y_test, yhat_nn_classes, average='macro')

print("F1 Score (Neural Network):", f1_nn)
```
